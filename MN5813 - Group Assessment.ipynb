### 3. Introduction

Over the past 120 years, the role and representation of women in the Olympics have gone through remarkable transformations, overcoming significant societal and institutional resistance. Pierre de Coubertin, the founder of the modern Olympic Games, although saying that sport open an opportunity for every man, woman and child, was openly resistant to the idea of women participating in the Olympics [(Wilant and Hirthler, 2024)](https://olympics.com/ioc/pierre-de-coubertin/the-rationale-behind-coubertins-opposition-to-women-competing-in-the-olympic-games). His belief that women’s participation would undermine the purpose and prestige of the Games reflects the deeply ingrained gender biases of the era. Despite his opposition, the inclusion of women in the 1900 Paris Olympics marked the beginning of a long journey toward gender equality in sports.

The article "The Changing Role of Women in the Olympic Games"[ (DeFrantz, 1997](https://www.bowlandmaths.org.uk/materials/projects/online/olympics/contents/The%20changing%20role%20of%20women%20in%20the%20Olympic%20games.pdf)) provides a comprehensive account of the challenges and milestones faced by women athletes. The articleopens upt on the controversies surrounding women’s participation in traditionally male-dominated sports. For example, the 1928 Amsterdam Games saw the inclusion of women’s track events, but backlash over perceived exhaustion led to the removal of the 800-meter race for decades. The article also underscores the broader cultural and political shifts that enabled greater inclusion of women in the Olympics, such as the rise of feminist movements and the recognition of sports as a tool for empowerment. By the end of the 20th century, the Olympics had moved closer to gender parity, with women participating in nearly all sports categories by the 2000 Games in Sydney.

This literature highlights the enduring struggle for women’s representation in the Olympics and provides a valuable context for analysing historical data. The progressive inclusion of women in the Games serves as a testament to the resilience of female athletes and advocates who challenged societal norms to achieve equality. This review forms the foundation for this report's exploration of the participation trends, age group distributions, and physical characteristics of female athletes over the last 120 years.

## III. Data Management

### 1. Data Cleaning

For missing data, the following tasks are executed: 

* **Drop NaN rows**: drop rows that does not contain any information
* **NaN in Medal column**: Fill null value with 'No Medal', in assumption that these athletes did not win any medal
* **NaN in region and notes columns**: Fill null value with 'No data' indicating the true status of these cells
* **NaN in Age, Height and Weight columns**: Fill null value with average number from each column, for the purpose of trend observations
  

```python
report_df.dropna()
report_df['Medal'] = report_df['Medal'].fillna('No Medal')
report_df[['region','notes']] = report_df[['region','notes']].fillna('No data')
report_df['Age'] = report_df['Age'].fillna(report_df['Age'].mean().round())
report_df['Height'] = report_df['Height'].fillna(report_df['Height'].mean().round())
report_df['Weight'] = report_df['Weight'].fillna(report_df['Weight'].mean().round())
```

### 1. Data Analysis

#### 1.1. Trends in Female and Male Athlete Participation Over Time

```python
# Create pivot table summarising female and male athlete for Summer Games only

summer_df =  report_df[report_df['Season'] == 'Summer']
gender_count_summer = summer_df.groupby(['Year', 'Sex'])['ID'].count().reset_index()

gender_count_summer_pivot = gender_count_summer.pivot_table(
    index='Year', 
    columns='Sex', 
    values='ID',
    fill_value=0,
).reset_index()

gender_count_summer_pivot = gender_count_summer_pivot.rename(columns={'F': 'Female_count', 'M': 'Male_count'})

gender_count_summer_pivot['%Growth_Female'] = (gender_count_summer_pivot['Female_count'].pct_change() * 100).round(2)
gender_count_summer_pivot['%Growth_Male'] = (gender_count_summer_pivot['Male_count'].pct_change() * 100).round(2)

# Create pivot table summarising female and male athlete for Winter Games only

winter_df =  report_df[report_df['Season'] == 'Winter']
gender_count_winter = winter_df.groupby(['Year', 'Sex'])['ID'].count().reset_index()

gender_count_winter_pivot = gender_count_winter.pivot_table(
    index='Year', 
    columns='Sex', 
    values='ID',
    fill_value=0,
).reset_index()

gender_count_winter_pivot = gender_count_winter_pivot.rename(columns={'F': 'Female_count', 'M': 'Male_count'})

gender_count_winter_pivot['%Growth_Female'] = (gender_count_winter_pivot['Female_count'].pct_change() * 100).round(2)
gender_count_winter_pivot['%Growth_Male'] = (gender_count_winter_pivot['Male_count'].pct_change() * 100).round(2)

# Combine both Winter and Summer table to create a comparison of participants for both seasons

combined_pivot = pd.concat(
    [gender_count_summer_pivot.set_index('Year'), gender_count_winter_pivot.set_index('Year')],
    axis=1, 
    keys=['Summer', 'Winter']
)

participation_trend = combined_pivot.replace([float('inf'), float('-inf')], np.nan).fillna(0).sort_index(axis=0, level=0, ascending=True)
participation_trend = participation_trend.reset_index()
```

#### 1.2. Female Athlete Age, Height and Weight by Sport

```python
# Creating new DataFrame contains Female athletes data only

women_only_df = report_df[report_df['Sex'] == 'F']


average_stats_women = (
    women_only_df.groupby('Sport')['Age']
    .agg(['mean'])
    .round(1)
)

# Extract first and latest heights for each sport
first_height = women_only_df.groupby('Sport')['Height'].first()
latest_height = women_only_df.groupby('Sport')['Height'].last()

# Extract first and latest weights for each sport
first_weight = women_only_df.groupby('Sport')['Weight'].first()
latest_weight = women_only_df.groupby('Sport')['Weight'].last()

# Add first and latest height to the stats DataFrame
average_stats_women[('Height', 'First Year')] = first_height
average_stats_women[('Height', 'Latest Year')] = latest_height

average_stats_women[('Weight', 'First Year')] = first_weight
average_stats_women[('Weight', 'Latest Year')] = latest_weight


# Add columns for difference in min and max of Height and Weight
average_stats_women[('Height', 'change')] = average_stats_women[('Height', 'Latest Year')] - average_stats_women[('Height', 'First Year')]
average_stats_women[('Weight', 'change')] = average_stats_women[('Weight', 'Latest Year')] - average_stats_women[('Weight', 'First Year')]
```

**Change in height vs change in weight over time across women’s sports**

```python
from scipy.stats import linregress

def calculate_changes(group):
    """Calculate changes in height and weight between first and last year for each sport"""
    first_year = group.loc[group['Year'].idxmin()]
    latest_year = group.loc[group['Year'].idxmax()]
    changes = {
        'Height Change': latest_year['Height'] - first_year['Height'],
        'Weight Change': latest_year['Weight'] - first_year['Weight']
    }
    return pd.Series(changes)

def analyze_athlete_changes(women_only_df):
    """Analyze and visualize changes in athlete measurements using hvplot"""
    # Calculate changes by sport with include_groups=False to handle deprecation warning
    changes_by_sport = women_only_df.groupby('Sport', as_index=False).apply(calculate_changes, include_groups=False)
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(
        changes_by_sport['Weight Change'],
        changes_by_sport['Height Change']
    )
    
    # Generate points for regression line
    x_range = changes_by_sport['Weight Change']
    x_line = np.linspace(min(x_range), max(x_range), 100)
    y_line = slope * x_line + intercept
    
    # Create regression line DataFrame
    regression_df = pd.DataFrame({
        'Weight Change': x_line,
        'Height Change': y_line
    })
    
    # Create scatter plot
    scatter_plot = changes_by_sport.hvplot.scatter(
        x='Weight Change',
        y='Height Change',
        by='Sport',
        title="Trends in Height vs Weight Changes of Female Athletes",
        height=600,
        width=1100,
        xlabel="Change in Weight (units)",
        ylabel="Change in Height (units)",
        legend='bottom_right'
    ).opts(legend_limit=10)
    
    # Create regression line plot
    regression_plot = regression_df.hvplot.line(
        x='Weight Change',
        y='Height Change',
        color='red',
        line_dash='dashed',
        label=f'Regression Line (slope={slope:.3f})'
    )
    
    # Combine plots
    combined_plot = scatter_plot * regression_plot
    
    stats = {
        'slope': slope,
        'intercept': intercept,
        'r_squared': r_value**2,
        'p_value': p_value
    }
    
    return combined_plot, stats
    
plot, stats = analyze_athlete_changes(women_only_df)

plot
```

**Discussion**

The chart chronicling the evolution of average athlete body measurements (weight and height) from 1900 to 2016 illustrates significant trends across various sports. Overall, most events have witnessed minimal changes in body metrics; in some cases, sports have evolved to favour lighter and shorter athletes.The one point that appears clearly in the upper right quadrant corresponds to the weightlifting, showcasing a trend toward heavier and taller competitors [(Karamanis, 2024)](https://www.visualcapitalist.com/cp/charted-olympic-athletes-getting-bigger/). Specifically, weightlifting stands out with an extraordinary increase of over 70 kg in average weight from the 1900 Olympics to the 2016 Rio Games. This highlights how changes in sports regulations, training techniques, and athletic specialization can influence the physical characteristics favoured in particular events.

Taken together, these charts highlight that the bodies of Olympic athletes have become increasingly extreme over time. While the trend is for athletes in most events to become taller and heavier, there are also a handful of events in which athletes have become smaller (e.g., Athletics) or simultaneously taller and lighter (e.g., tennis).
